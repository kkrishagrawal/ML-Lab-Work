```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'admission - admission.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#Bar plot
plt.bar(df['University Rating'], df['Chance of Admit'], color='blue')
plt.xlabel('University Rating')
plt.ylabel('Chance of admit')
plt.title('University Rating VS Chance of admit')
plt.show()
```

```python
#Scatter plot
plt.scatter(df['GRE Score'], df['CGPA'])
plt.xlabel('GRE Score')
plt.ylabel('CGPA')
plt.title('Relation between GRE Score and CGPA')
plt.show()
```

```python
#Bar plot
plt.bar(df['Chance of Admit'], df['LOR'], color='blue')
plt.ylabel('LOR')
plt.xlabel('Chance of admit')
plt.title('LOR VS Chance of admit')
plt.show()
```

```python
#Box plot
sns.boxplot(x = "CGPA", data = df, showmeans="True", color = "blue")
plt.title('Box Plot of CGPA')
```

```python
#Box plot
plt.figure(figsize=(6,4))
sns.boxplot(x=df['University Rating'], y=df['TOEFL Score'], color = 'blue')
plt.title('University Rating vs TOEFL Score')
plt.show()
```

```python
#Line chart
plt.plot(df["Serial No."], df["CGPA"], color = 'blue')
plt.xlabel('Serial No.')
plt.ylabel('CGPA')
plt.title('')
plt.show()
```
